# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\juliu\Documents\Work\Python\Pandas\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('seaborn-v0_8-deep')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar', stacked = True, title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
df.plot.barh(stacked = True, title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'Green')
```

```python
df.plot.hist(bins = 20)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (20,10))
```
